```python
import pandas as pd

# Fetch event records
cursor.execute("SELECT id, title, description, tags FROM events")
rows = cursor.fetchall()

# Create DataFrame
df = pd.DataFrame(rows, columns=["id", "title", "description", "tags"])

# Combine title + description + tags into 1 string
def merge_event_fields(row):
    tag_str = ' '.join(row['tags']) if isinstance(row['tags'], list) else ''
    return f"{row['title']} {row['description']} {tag_str}"

df['combined'] = df.apply(merge_event_fields, axis=1)

# Display
print("✅ Events fetched and processed.")
df.head()
```

```python
!pip install -q sentence-transformers

from sentence_transformers import SentenceTransformer

# Load the model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Create embedding vectors for all events
embeddings = model.encode(
    df['combined'].tolist(),       # List of combined event strings
    show_progress_bar=True         # Show loading bar
)

print(f"✅ Generated {len(embeddings)} embeddings of size {len(embeddings[0])}")
```

```python
print("Text:", df['combined'][0])
print("Vector:", embeddings[0])
```

```python
import faiss
import numpy as np

# Convert to float32 (required by FAISS)
embedding_matrix = np.array(embeddings).astype('float32')

# Dimensions must match your model’s output (e.g., 384 for MiniLM)
embedding_dim = embedding_matrix.shape[1]

# 🔄 Normalize vectors to unit length (needed for cosine similarity)
faiss.normalize_L2(embedding_matrix)

# ✅ Create FAISS index using Inner Product
index = faiss.IndexFlatIP(embedding_dim)

# 💾 Add normalized embeddings only once
index.add(embedding_matrix)

print(f"✅ FAISS index built with {index.ntotal} vectors.")
```

```python
# Example: get 5 most similar events to the first one
query_vector = embedding_matrix[0].reshape(1, -1)
top_k = 5
distances, indices = index.search(query_vector, top_k)

print("Similar events to:", df['title'][0])
for i in range(top_k):
    idx = indices[0][i]
    print(f" - {df['title'][idx]} (Distance: {distances[0][i]:.4f})")
```

```python
from supabase import create_client

# Use your project's credentials
SUPABASE_URL = "add your own"
SUPABASE_KEY = "add your own"

supabase = create_client(SUPABASE_URL, SUPABASE_KEY)

# Fetch users and their interests (join logic done in Python)
users = supabase.table("users").select("id, full_name").execute().data
interests = supabase.table("user_interests").select("user_id, interests").execute().data

# Convert to DataFrame and merge
users_df = pd.DataFrame(users)
interests_df = pd.DataFrame(interests)
user_df = pd.merge(users_df, interests_df, left_on="id", right_on="user_id")

print(user_df.head())
```

```python
from sentence_transformers import SentenceTransformer

# Load the same embedding model used earlier
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# Create synthetic "interest sentences" for each user
user_df["interest_text"] = user_df["interests"].apply(lambda tags: " ".join(tags))

# Generate embeddings
user_embeddings = model.encode(user_df["interest_text"].tolist(), show_progress_bar=True)

print("✅ Generated cold-start embeddings for users")
```

```python
# Ensure event FAISS index is already built and contains normalized event embeddings
# (You should already have this from before)
assert index.ntotal == len(df), "FAISS index is not ready!"

top_k = 5

# Run similarity search for each user
for user_idx, user_vector in enumerate(user_embeddings):
    # Normalize the user vector for cosine similarity
    user_vector = user_vector.reshape(1, -1)
    faiss.normalize_L2(user_vector)

    # Search the index
    distances, indices = index.search(user_vector, top_k)

    user_name = user_df.iloc[user_idx]["full_name"]
    print(f"\n🎯 Recommendations for {user_name}:")

    for i in range(top_k):
        event_idx = indices[0][i]
        event_title = df.iloc[event_idx]["title"]
        similarity_score = distances[0][i]
        print(f"  {i+1}. {event_title} (Score: {similarity_score:.4f})")
```
